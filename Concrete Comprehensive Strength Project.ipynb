```python
for feature in continous_feature:
    data = df.copy()
    bar =sb.distplot(data[feature])
    bar.legend(["Skewness: {:.2f}".format(data[feature].skew())])
    plt.title(feature)
    plt.xlabel(feature, color='r')
    plt.show()
```

```python
over_skewed = []
for x in continous_feature:
    if df[x].skew()>0.50:
        over_skewed.append(x)
```

```python
over_skewed
```

```python
for feature in over_skewed:
    data = df.copy()
    data[feature] = np.log1p(data[feature])
    bar =sb.distplot(data[feature])
    bar.legend(["Skewness: {:.2f}".format(data[feature].skew())])
    plt.title(feature)
    plt.xlabel(feature, color='r')
    plt.show()
```

```python
corr_ftr = df.corr()
high_corr = []
```

```python
y_tr = df.iloc[:,0:8]
X = df.iloc[:,-1]
```

```python
y_tr.head(2)
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(y_tr,X, train_size = 0.75, random_state = 5)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
pred = lr.predict(X_test)
```

```python
plt.scatter(pred,Y_test)
plt.title("Predicted values vs Actual values")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
```

```python
acc = r2_score(Y_test,pred)
```

```python
acc
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(X_train,Y_train)
```

```python
predrf = rf.predict(X_test)
```

```python
acc_rf = r2_score(Y_test,predrf)
```

```python
rmse = np.sqrt(mean_squared_error(Y_test, pred))
```
